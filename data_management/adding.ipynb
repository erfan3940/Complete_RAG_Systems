```python
import re
class PDFProcessor :
    """complete pdf processing for embedding"""
    def __init__(self,chunk_size=500,chunk_overlap=100):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size = chunk_size,
            chunk_overlap = chunk_overlap,
            separators = ["\n"," "]
        )
    def preprocess (self,path) -> List[Document]:
        """loading to chunking"""
        # load data
        try:
            loader = PyMuPDFLoader(path)
            loaded = loader.load()
            print("loaded succesfully")
        except Exception as e :
            print(f"error is {e}")

        # spltting
        final_chunk = []
        for page_num,page in enumerate (loaded):
            # clean text
            cleaned_text = self._clean_text(page.page_content)
            # pass empty pages
            if len(cleaned_text) < 50:
                print("is less than 50")
                continue
            # chunking
            created = self.text_splitter.create_documents(
                texts=[cleaned_text],
                metadatas=[{
                    **page.metadata,
                    "page" : page_num + 1,
                    "total_pages" : len(loaded),
                    "method" : "PDFProccessor class",
                    "word_count": len(cleaned_text.split()),
                    "char_count": len(cleaned_text),
                    "source": path
                }])
            final_chunk.extend(created)
            
        print(f"Created {len(final_chunk)} chunks from {len(loaded)} pages")
        return final_chunk
        
        
    def _clean_text (self,text) -> str:
        # text = " ".join(text.strip())
        # Remove excessive whitespace
        text = re.sub(r'\s+', ' ', text)
        
        # Remove special PDF artifacts
        text = re.sub(r'\x0c', '', text)  # Form feed
        text = re.sub(r'\ufeff', '', text)  # BOM
        text = re.sub(r'\x00', '', text)  # Null bytes
        
        # Normalize line endings
        text = text.replace('\r\n', '\n').replace('\r', '\n')
        
        # Remove non-ASCII characters (optional)
        # text = text.encode('ascii', 'ignore').decode('ascii')
        
        # Strip leading/trailing whitespace
        text = text.strip()
        
        return text

        



```

```python
from psycopg2.extras import RealDictCursor
cursor = connection.cursor(cursor_factory=RealDictCursor)
cursor.execute("""SELECT e.id AS employee_id,f_name,l_name,i.email,password,p.id AS project_id,p.project_name,team_id,deadline,t.team_name,c_name FROM employees AS e
LEFT JOIN interanet_accounts AS i ON i.email = e.email
LEFT JOIN projects_employees AS pe ON e.id = pe.employee_id
LEFT JOIN projects AS p ON p.id = pe.project_id
LEFT JOIN teams AS t ON t.id = e.team_id AND t.id = 1 
LEFT JOIN company AS c ON c.id = t.building_id ;""")
rows = cursor.fetchall()
```

```python
from langchain_core.documents import Document
documents = []
for row in rows:
    text = (
        f"Employee ID: {row['employee_id']} "
        f"Name: {row['f_name']} {row['l_name']} "
        f"email: {row['email']} "
        f"project_name: {row['project_name']} "
        f"project_id: {row['project_id']} "
        f"deadline: {row['deadline']} "
    )

    doc = Document(
        page_content=text,
        metadata={
            "employee_id": row["employee_id"],
            "query": "all join"
        }
    )

    documents.append(doc)
```
